### Ordenar por ranking e salvar os 1000 mais bem rankeados em um arquivo CSV
##### Print na tela de 5 linhas

```
#LIMPANDO OS VALORES ERRADOS DO DATA FRAME E FILTRANDO A PARTIR DOS ANOS 2000
#limpando a coluna ano com valor em string
movies_csv['year'] = movies_csv['year'].apply(pd.to_numeric, errors='coerce')
movies_csv['year'] = movies_csv['year'].dropna()
movies_csv

#selecionando os filmes a partir dos anos 2000
movies_csv = movies_csv.loc[(movies_csv['year'] >= 2000)]
movies_csv.head()
```

```
###MÉDIA E CRIAÇÃO DO RANKING
#selecionando colunas
colunas = movies_csv.loc[:, 'avg_vote':'votes']
#tirando a média entre elas
movies_csv['votesMean'] = colunas.mean(axis=1)
#colocando em ordem decrescente de acordo com a coluna de média e selecionando os 1000 filmes mais rankeados
best_movies = movies_csv.sort_values(by='votesMean', ascending=False)[:1000]
best_movies.head()
```
