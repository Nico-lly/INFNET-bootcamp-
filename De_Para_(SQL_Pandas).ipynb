```
Cr_header = ['StudentId', 'CourseId','Grade','Attempt']
Credit = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL exercicio/SampleCreditData', names = Cr_header)
Credit.head()
```

```
#Estudantes média acima de 83.5

#Dados de todos os estudantes com média acima de 83.5
StudentGrade = pd.merge(Student,Credit, how='inner')
HScore = StudentGrade[StudentGrade['Grade'] > 83.5]
HScore.head()

#Quantidade de estudantes com média acima de 83.5
Qnt = Credit['StudentId'][Credit['Grade']> 83.5].count()
print(Qnt)
```

```
#Usando uma lógica condicional retorne : Id do aluno, nota, aprovado ou reprovado. 
#Aprovado se nota >= 70.0 Dica: Usar o método pandas.assign com o método np.where

import numpy as np
```

```
StudentGrade['Aproved'] = np.where(StudentGrade['Grade']>=70, 'Aprovado', 'Reprovado')
#Dois colchetes pois o acesso é a uma lista de colunas (mais de um valor)
StudentGrade[['StudentId', 'Grade', 'Aproved']]
```

```
#Resolvendo só com pandas e função lambda
Credit['Aproved'] = Credit.apply(lambda x: 'aprovado' if x.Grade >= 70 else 'reprovado', axis=1)
Credit[['StudentId', 'Grade', 'Aproved']]
```

```
#Retorne um dataframe com os nomes dos cursos cuja a média obtida (AVG) pelo alunos foi maior ou igual 75.0

CourseAVG = Credit.groupby('CourseId')['Grade'].mean()
CourseGrade = Course.merge(CourseAVG[CourseAVG >= 60], on = 'CourseId').sort_values('Grade', ascending = False)
CourseGrade.head()
```

```
#Retorne um datafrema com os email dos alunos que cursaram 3 vezes um mesmo curso
SameStudent = Credit[Credit['Attempt']==3]
A = SameStudent.merge(Student, how = 'inner', on = 'StudentId')['Email']
A.head()
```
