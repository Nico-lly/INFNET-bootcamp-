##Carregar os dados dos arquivos nas respectivas tabelas

```
import sqlite3

#Criando o banco de dados que vai receber todas as tabelas
bd = sqlite3.connect('universidade.db')
```

```
#Função que falta no sqlite3 para imprimir os títulos das linhas 
def consulta(sql):
  data = cursor.execute(sql)
  header =[header[0] for header in data.description]
  print(header)
  for row in data:
    print(row)
```

```
#criando um cursor para se movimentar no banco de dados usando o sqlite3
cursor = bd.cursor()
```

```
#Criar a tabela PERSON no banco de dados
SQL = """CREATE TABLE IF NOT EXISTS person(
          Personid integer PRIMARY KEY,
          FirstName text,
          MiddleInitial text,
          LastName text,
          DateOfBirth text
          );"""
```

```
#Executando a criação da tabela
cursor.execute(SQL)
```

```
#Comitar a tabela criada no banco de dados
bd.commit()
```

```
#Verificar se a tabela PERSONS foi criada corretamente
cursor.execute("PRAGMA table_info(person)").fetchall()
```

```
####Inserir arquivo na tabela criada

import csv

#abrindo e lendo o arquivo
csv_file = open('/content/drive/MyDrive/Colab Notebooks/SQL exercicio/SamplePersonData')
Persons = csv.reader(csv_file)
```

```
#Inserindo os dados do arquivo lido na tabela criada no banco de dados
cursor.executemany('INSERT INTO person VALUES(?, ?, ?, ?, ?)', Persons )
```

```
#Comitando as informações carregadas no banco de dados
bd.commit()
```

```
#Verificando se os dados foram inseridos
consulta('SELECT * FROM person')
```
